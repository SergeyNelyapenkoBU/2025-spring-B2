Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

#### FIGURE 4.8  Leverage color to draw attention (page 111)

```python
# Sample data (not exact values from the book)
concerns = [
    "Engine power is less than expected",
    "Tires make excessive noise while driving",
    "Engine makes abnormal/excessive noise",
    "Seat material concerns",
    "Excessive wind noise",
    "Hesitation or delay when shifting",
    "Bluetooth system has poor sound quality",
    "Steering system/wheel has too much play",
    "Bluetooth connectivity issues",
    "Front seat audio/navigation controls"
]
values = [12.9, 12.6, 11.6, 11.1, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]

# Decide which bars to highlight (e.g., top 6 in red, rest in gray)
colors = ["#d6604d"] * 6 + ["#7f7f7f"] * 4  # red for top 6, gray for last 4

# Sort data by value in descending order so largest bars are at the top
sorted_data = sorted(zip(concerns, values, colors), key=lambda x: x[1])
concerns_sorted, values_sorted, colors_sorted = zip(*sorted_data)

# Create figure and axis
fig, ax = plt.subplots(figsize=(7, 5))

# Plot horizontal bars
bars = ax.barh(concerns_sorted, values_sorted, color=colors_sorted)

# Remove x-axis and y-axis spines for a cleaner look
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# Remove gridlines
ax.grid(False)

# No need to invert the y-axis since data is already sorted descending
# ax.invert_yaxis()  <-- Removed

# Add data labels on the right side of each bar
for bar in bars:
    width = bar.get_width()
    y_pos = bar.get_y() + bar.get_height() / 2
    ax.text(width + 0.3, y_pos, f"{width:.1f}", va="center")

# Label the x-axis (optional, can be subtle)
ax.set_xlabel("concerns per 1,000", labelpad=10)
ax.xaxis.set_label_position('top')
# plt.xlabel("concerns per 1,000", fontsize=12)

# Add a title that matches the example's context
#ax.set_title("Top 10 design concerns", pad=-20)
plt.figtext(0, 0.95, "Top 10 design concerns", ha='left', fontsize=12)
# Caption below the plot
plt.figtext(0, -0.04, "FIGURE 4.8  Leverage color to draw attention", ha='left', fontsize=10)

plt.tight_layout()
plt.show()

```
