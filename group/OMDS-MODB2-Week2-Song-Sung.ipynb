```python
# download kaggle file
#!pip install kagglehub
import kagglehub
vaccine_dataset_path = kagglehub.dataset_download("deepshah16/vaccine-hesitancy-for-covid19")
print("Path to dataset file:", vaccine_dataset_path)
import os
vaccine_dataset_path_to_file = os.path.join(vaccine_dataset_path, "Vaccine_Hesitancy_for_COVID-19__County_and_local_estimates.csv")
vaccine_dataset_raw = pd.read_csv(vaccine_dataset_path_to_file)
#print(vaccine_dataset_raw.head())
vaccine_dataset_raw.head()
```

```python
# summarize the datasets using info() and describe()

print("info")
vaccine_dataset_raw.info()

print("describe")
vaccine_dataset_raw.describe()
```

```python
# Are there any duplicate rows?
duplicates = vaccine_dataset_raw.duplicated()
number_duplicates = duplicates.sum()

print(f'Number of duplicate rows: {number_duplicates}')
if number_duplicates == 0:
    print("No duplicate rows found.")
else:
    print(f"{number_duplicates} duplicate rows found.")
```

```python
# Are there any duplicate values in a given column (when this would be inappropriate?)

column_names = vaccine_dataset_raw.columns.tolist()

for column in column_names:
    duplicates_column = vaccine_dataset_raw[column].duplicated()
    number_duplicates_column = duplicates_column.sum()

    if number_duplicates_column == 0:
        print(f"No duplicate values found in the column '{column}'.")
    else:
        print(f"{number_duplicates_column} duplicate values found in the column '{column}'.")
```

Any inappropriate duplicate values in a given column.
Geographical point should be unique. Duplicates may mean redundancy in location.
State code and state duplicates is ok because it is looking at more detailed level like county name.
county boundary and state boundary should be unique for each county and each state. Duplicates could mean overlapping or redundant boundries.



```python
print("What are the mean, median, and mode of each numerical column?")
from scipy import stats
numeric_columns = [
    'Estimated hesitant',
    'Estimated strongly hesitant',
    'Social Vulnerability Index (SVI)',
    'Percent adults fully vaccinated against COVID-19',
    'Percent Hispanic',
    'Percent non-Hispanic American Indian/Alaska Native',
    'Percent non-Hispanic Asian',
    'Percent non-Hispanic Black',
    'Percent non-Hispanic Native Hawaiian/Pacific Islander',
    'Percent non-Hispanic White',
    'CVAC level of concern for vaccination rollout'
]

# Calculate mean, median, and mode for each numeric column
results = {}
for col in numeric_columns:
    mean_value = vaccine_dataset_raw[col].mean()
    median_value = vaccine_dataset_raw[col].median()
    # Handle NaN values
    mode_result = stats.mode(vaccine_dataset_raw[col].dropna(), nan_policy='omit', keepdims=True)
    mode_value = mode_result.mode[0] if mode_result.mode.size > 0 else None
    results[col] = {
        'Mean': mean_value,
        'Median': median_value,
        'Mode': mode_value
    }

summary_df = pd.DataFrame(results).T

print(summary_df)

print("categorical to calculate mode")
categorical_columns = [
    'FIPS Code', 'County Name', 'State', 'SVI Category',
    'CVAC Level Of Concern',
    'Geographical Point', 'State Code', 'County Boundary', 'State Boundary'
]
categorical_mode_values = {}
for col in categorical_columns:
    if col in vaccine_dataset_raw.columns:
        # Drop NaN values before calculating mode
        mode_value = vaccine_dataset_raw[col].dropna().mode()
        if not mode_value.empty:
            categorical_mode_values[col] = mode_value[0]
        else:
            categorical_mode_values[col] = None

for col, mode in categorical_mode_values.items():
    print(f"Mode of {col}: {mode}")
```

```python
print("Are there any missing or null values? Handle them.")
print("Just by filtering in the csv file, there are some columns with missing values in some cells.")

print("Handle missing values in numeric columns: fill in with mean")
for col in numeric_columns:
    if col in vaccine_dataset_raw.columns:
        #Fill missing values in numeric columns with the mean
        vaccine_dataset_raw[col] = vaccine_dataset_raw[col].fillna(vaccine_dataset_raw[col].mean())

print("Handle missing values in categorical columns: fill in with mode")
for col in categorical_columns:
    if col in vaccine_dataset_raw.columns:
        #Fill missing values in categorical columns with the mode
        mode_value = vaccine_dataset_raw[col].dropna().mode()
        if not mode_value.empty:
            vaccine_dataset_raw[col] = vaccine_dataset_raw[col].fillna(mode_value[0])
```

```python
print("Identify any other inconsistent data (e.g. someone seems to be taking an action before they are born.")
print("Looking through the data, there are no other inconsistencies to be dealt with in the dataset.")
```

```python
print("Encode any categorical variables (e.g. with one-hot encoding)")
hot_encoded_dataset = pd.get_dummies(vaccine_dataset_raw, columns=categorical_columns, drop_first=True)

print("One-hot encoded dataset:")
print(hot_encoded_dataset.head().T)
```

```python
import matplotlib.pyplot as plt

# Data for the stacked bar chart
categories = ['US Population', 'Our Customers']
segments = ['Segment 1', 'Segment 2', 'Segment 3', 'Segment 4', 'Segment 5', 'Segment 6', 'Segment 7']
colors = ['red', 'darkgreen', 'navy', 'grey', 'lightblue', 'yellow', 'lightgreen']  # Adjusted Segment 7 to light green

us_population = [16, 7, 10, 10, 10, 32, 15]  # Segment 1 at the bottom, Segment 7 at the top
our_customers = [9, 10, 15, 18, 17, 20, 11]

fig, ax = plt.subplots(figsize=(8, 6))

bottoms = [0, 0]  
for i, (segment, color) in enumerate(zip(segments, colors)):
    heights = [us_population[i], our_customers[i]]
    ax.bar(categories, heights, bottom=bottoms, color=color, edgecolor='black')
    bottoms = [bottoms[0] + heights[0], bottoms[1] + heights[1]]

for i, segment in enumerate(segments):
    label_color = 'black'
    ax.text(-0.5, sum(us_population[:i]) + us_population[i] / 2, segment, 
            va='center', ha='right', fontsize=10, color=label_color)

for category_idx in range(len(categories)):
    cumulative_height = 0
    for i, height in enumerate([us_population, our_customers][category_idx]):
        cumulative_height += height
        ax.text(category_idx, cumulative_height - height / 2, f'{height}%', 
                ha='center', va='center', fontsize=9, color='black')

ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_yticks([])

ax.set_title('Our Customers', fontsize=14, weight='bold', pad=-20)
ax.set_ylabel('')
ax.set_xlabel('')

plt.tight_layout()
plt.show()

```
