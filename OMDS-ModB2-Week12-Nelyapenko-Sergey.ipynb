# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

Advantages:
  - Clarity of individual trends: Each week is isolated, so it’s easier to analyze the visitor trend within each week independently.
  - Less visual clutter: Especially helpful if the data is noisy or overlapping.
  - Labeling flexibility: You can tailor axis ranges and labels per subplot if needed.

Disadvantages:
  - Harder to compare weeks directly: You have to visually jump between subplots to compare same-day values.
  - Space-consuming: Takes more screen or print space, which can be limiting for dashboards or reports.

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

Advantages:
  - Easy comparison: You can immediately see differences and similarities between the two weeks — e.g., if visitor numbers on Thursday were higher in Week 2 than Week 1.
  - More compact presentation: Saves space and is suitable for quick side-by-side comparison.

Disadvantages:
  - Can become cluttered: Especially if more weeks or datasets are added — overlapping lines can reduce clarity.
  - Harder to interpret individual weeks: If the lines cross frequently or the color distinction isn’t clear, it may confuse viewers.


Best Use Cases:
  - Use separate subplots when you want to focus on the details of each week.
  - Use overlayed lines when comparing trends or performance between weeks is the priority.

```python
import matplotlib.pyplot as plt

years = list(range(2008, 2015))

product_a = [400, 420, 430, 410, 370, 300, 290]
product_b = [380, 400, 420, 400, 360, 290, 280]
product_c = [100, 150, 200, 220, 230, 210, 200]
product_d = [0, 0, 0, 200, 250, 200, 190]
product_e = [0, 0, 0, 0, 0, 150, 230]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(years, product_a, label="Product A", color='blue')
plt.plot(years, product_b, label="Product B", color='brown')
plt.plot(years, product_c, label="Product C", color='green')
plt.plot(years, product_d, label="Product D", color='purple')
plt.plot(years, product_e, label="Product E", color='cyan')

# Title and axis labels
plt.title("Average Retail Product Price per Year", fontsize=14)
plt.ylabel("$0", rotation=0, labelpad=40)
plt.ylim(0, 500)
plt.xticks(years)
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

# Legend at the bottom
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=5)

plt.tight_layout()
plt.show()
```
